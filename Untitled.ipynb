```python
import pandas as pd
import re
import os
import mmap

'''Read and transform data'''

data = pd.read_excel("Numbers/Phone_Numbers.xlsx")
data=data.T.reset_index()
df=pd.Series(data.values.ravel("C"))
df.shape
```

```python
# Remove null values
df=df.dropna()
df.shape
```

```python
data = []
```

```python
ken = re.compile("254[17]")
for i in pd.Series.iteritems(df):
    i=str(i[1])
#     i=re.sub(' ','',i)
    i=re.sub('[A-Z a-z(-)-+=":_"<>/]','',i)
    i=i.split(".")[0]
    if len(i) < 9:
        pass
    elif len(i) > 14:
        i= i[0:13]
    elif ken.fullmatch(i, 0,4) != None:
        pass
    else:
        data.append(i)
```

```python
len(data)
```

```python
from collections import Counter
counts = dict(Counter(data))
```

```python
sort = sorted(counts.items(), key=lambda x:x[1], reverse=True)

with open(os.path.join("Numbers", "Employers2.txt"), 'a+', encoding="utf-8") as f:
    lines = f.readlines()
    for i in sort:
        if i[1] > 3:
            if lines.find(i[0]) != -1:
#             if i[0] in f.read():
                pass
            else:
                f.write(f"{i[0]}\n")
                f.close()
        else:
            print(i[0], i[1])
```
